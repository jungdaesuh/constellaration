# Sample and visualize Omnigenous Poloidal fields

```python
from constellaration.omnigeneity import omnigenity_field_sampling
from constellaration.omnigeneity import omnigenity_field_plot
import matplotlib.pyplot as plt
```

```python
# Create sampling settings with default parameters
settings = omnigenity_field_sampling.SampleOmnigenousFieldAndTargetsSettings(n_samples=5)

# Generate sample fields and targets
samples = settings.sample_omnigenous_fields_and_targets()

# Inspect the first sample
print(samples[0].model_dump_json(indent=4))
```

## Visualizing an Omnigenous Poloidal field

```python
# Visualize the field

omnigenity_field_plot.plot_boozer_field(samples[0].omnigenous_field);
```

```python
# Visualize the magnetic well
fig = omnigenity_field_plot.plot_modb_well(samples[0].omnigenous_field)
fig
```

```python
# Visualize many samples

settings = omnigenity_field_sampling.SampleOmnigenousFieldAndTargetsSettings(
    n_samples=30, seed=42)
samples = settings.sample_omnigenous_fields_and_targets()


fig, axes = plt.subplots(6, 5, figsize=(18, 12))
for sample, ax in zip(samples, axes.flatten()):
    omnigenity_field_plot.plot_boozer_field(
        sample.omnigenous_field, ax=ax, show=False)
```
